```python
%matplotlib inline

import numpy as np
import matplotlib
import matplotlib.image as image
import matplotlib.pyplot as plt
from skimage.feature import (corner_harris, corner_peaks, plot_matches, BRIEF, match_descriptors)
from skimage.transform import warp, ProjectiveTransform
from skimage.color import rgb2gray
from skimage.measure import ransac
```

### feature detection using corners and feature matching

```python
#imL = image.imread("images/CMU_left.jpg")
#imR = image.imread("images/CMU_right.jpg")
imL = image.imread("images/left.jpg")
imR = image.imread("images/right.jpg")
imLgray = rgb2gray(imL)
imRgray = rgb2gray(imR)

keypointsL = corner_peaks(corner_harris(imLgray), threshold_rel=0.0005, min_distance=5)
keypointsR = corner_peaks(corner_harris(imRgray), threshold_rel=0.0005, min_distance=5)

extractor = BRIEF()

extractor.extract(imLgray, keypointsL)
keypointsL = keypointsL[extractor.mask]         
descriptorsL = extractor.descriptors

extractor.extract(imRgray, keypointsR)
keypointsR = keypointsR[extractor.mask]
descriptorsR = extractor.descriptors

matchesLR = match_descriptors(descriptorsL, descriptorsR, cross_check=True)

fig = plt.figure(1,figsize = (12, 4))
axA = plt.subplot(111)
plt.gray()
plot_matches(axA, imL, imR, keypointsL, keypointsR, matchesLR) #, matches_color = 'r')
axA.axis('off')

plt.show()
```

### using RANSAC to estimate Homography from matched pairs of points above. This cell display new figure 2 similar to figure 1 above, but it should show only inlier pairs for the detected homography.

```python
fig = plt.figure(2,figsize = (12, 14))
axA = plt.subplot(111)
plt.gray()
plot_matches(axA, imL, imR, keypointsL, keypointsR, matchesLR) #, matches_color = 'r')
axA.axis('off')
plt.show()

# robustly fit line using RANSAC algorithm
model_robust, inliers = ransac((keypointsL[matchesLR[:, 0]],keypointsR[matchesLR[:, 1]]), ProjectiveTransform, min_samples=4, residual_threshold=5, max_trials=120)
outliers = inliers == False

#plot the result
fig = plt.figure(10,figsize = (12, 14))
axB = plt.subplot(111)
plt.gray()
inlier_idxs = np.nonzero(inliers)[0]
plot_matches(axB, imL, imR, keypointsL[matchesLR[:, 0]], keypointsR[matchesLR[:, 1]],
             np.column_stack((inlier_idxs, inlier_idxs)))
axB.axis('off')
plt.show()
```

## reprojecting onto common projection plane
### show the reference image (left) and projected right image 

```python
from numpy import linalg
fig = plt.figure(3,figsize = (12, 14))
plt.subplot(311)
I = [[1,0,0],[0,1,0],[0,0,1]]
Ir = [[1,0,-350],[0,1,0],[0,0,1]]
I = np.array(I)
Ir = np.array(Ir)
left_image = warp(imL, I, output_shape=(300, 700))
right_image = warp(imR, Ir, output_shape=(300, 700))
plt.imshow(left_image)
plt.title("reference frame with the left image")
plt.subplot(312)
#warped = warp(imL, ProjectiveTransform(matrix=matrix))

p1 = keypointsL[matchesLR[:, 0]][inliers]
p2 = keypointsR[matchesLR[:, 1]][inliers]
#construct matrix a
A = []
for i in range(p1[:,0].size):
    A.append([p1[i][1] , p1[i][0], 1, 0, 0, 0, -p1[i][1]*p2[i][1],-p1[i][0]*p2[i][1], -p2[i][1] ])
    A.append([0, 0, 0, p1[i][1] , p1[i][0], 1, -p1[i][1]*p2[i][0],-p1[i][0]*p2[i][0], -p2[i][0] ])
a = np.array(A)
# find transformation h
A_T_A = np.dot(a.T, a)
eigen_values, eigen_vectors = linalg.eig(A_T_A)
h = eigen_vectors[:,eigen_values.argmin()]
h = h.reshape(3,3)
# warp the image
warped1 = warp(imR,h,output_shape=(300, 700),mode='constant')
plt.imshow(warped1)
plt.title("reference frame with the right image (reprojected)")

```

## blending and the final result

```python
'''
    This function compute distance transform for the boundary of a given image.
    Return a numpy array of the same size as the image with distances from each pixel 
    to the closest point on the boundary of the image (float values).
'''
def boundaryDT(image):
    w = image.shape[0]
    h = image.shape[1]
    result = np.arange(w*h).reshape(w,h)
    for i in range(w):
        for j in range(h):
            result[i][j] = min(i,j,w-i,h-j)
    return result

'''
    Compute a distance transform for both images(using boundaryDT)
    Show the result of both reference img and projected right img
'''
fig = plt.figure(4,figsize = (12, 3))
plt.subplot(121)
LboundaryDT = boundaryDT(imL)
dtrans1 = warp(LboundaryDT, I, output_shape=(300, 700))
plt.imshow(dtrans1)
plt.title("Left image DT in Ref. frame (LdtRef)")

plt.subplot(122)
Rboundary = boundaryDT(imR)
dtrans2 = warp(Rboundary,h,output_shape=(300, 700),mode='constant')
plt.imshow(dtrans2)
plt.title("Right image DT in Ref. frame (RdtRef)")
```

```python
'''
    Use boundary distance transforms to blend left and right images (reprojected) into the reference frame. 
'''
# to avoid divided by 0, we calculate alpha1 using loop
alpha1 = dtrans1.copy()
for i in range(dtrans1.shape[0]):
    for j in range(dtrans1.shape[1]):
        if(dtrans1[i][j] == 0 and dtrans2[i][j] == 0):
            alpha1[i][j] = 0
        else: 
            alpha1[i][j]= dtrans1[i][j] / (dtrans1[i][j] + dtrans2[i][j])
# the first image
fig = plt.figure(5,figsize = (12, 14))
plt.subplot(311)
R1 = left_image[:,:,0]
G1 = left_image[:,:,1] 
B1 = left_image[:,:,2]
R1_blend = alpha1 * R1
G1_blend = alpha1 * G1 
B1_blend = alpha1 * B1
Lblended = np.dstack((R1_blend,G1_blend,B1_blend))
plt.imshow(Lblended)
plt.title("alpha based on DT (for RransacRef)")

# The second plot
plt.subplot(312)
alpha2 = 1- alpha1
R2 = warped1[:,:,0]  
G2 = warped1[:,:,1]  
B2 = warped1[:,:,2]
R2_blend = alpha2 * R2
G2_blend = alpha2 * G2 
B2_blend = alpha2 * B2
Rblended = np.dstack((R2_blend,G2_blend,B2_blend))
plt.imshow(Rblended)
plt.title("alpha based on DT (for RransacRef)")

# the third plot
plt.subplot(313)
R_blend = R1_blend + R2_blend
G_blend = G1_blend + G2_blend
B_blend = B1_blend + B2_blend
final_blended = np.dstack((R_blend,G_blend,B_blend))

plt.imshow(final_blended)
plt.title("Panorama")

plt.show()
```
